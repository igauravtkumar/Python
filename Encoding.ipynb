```python
# Encoding --> Transform categorical data into Numerical data.

# 1> Label Encoding --> apply on ingle column and here we can not decide for
#  which sub-category what will be the number

# 2> Ordinal Encoding --> apply on multiple columns and here decide sub-category value.
# 3> OneHot Encoding --> it will create new column based on the sub-category
# 4> get_dummies --> It will create new column based on the sub-categories
```

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/covid_toy.csv')
df.head(3)
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
df.isnull().sum()
```

```python
df.head(3)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lb = LabelEncoder()
```

```python
df['gender']=lb.fit_transform(df['gender'])
df.head(3)
```

```python
df['cough']=lb.fit_transform(df['cough'])
df['city']=lb.fit_transform(df['city'])
df['has_covid']=lb.fit_transform(df['has_covid'])
```

```python
df.sample(5)
```

```python
# OneHotEncoding
```

```python
df = pd.read_csv('/content/covid_toy.csv')
df.head(2)
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(drop = 'first', sparse_output= False , dtype = np.int32)
```

```python
df_new = ohe.fit_transform(df[['gender','cough','city','has_covid']])
```

```python
df_new
```

```python
# get_dummies -->
```

```python
a = pd.get_dummies(df, columns = ['gender','cough','city','has_covid'], drop_first = True)
# drop_first --> first column ki subcategory ko remove
```

```python
a=a.dropna()
a.head(3)
```

```python
a = a.astype(int)
a
```

```python
from sklearn.model_selection import train_test_split
```

```python
x = a.drop(columns = ['has_covid_Yes'])
y=a['has_covid_Yes']
```

```python
x_train ,  x_test , y_train , y_test = train_test_split(x,
                                                        y,
                                                        test_size = 0.2,
                                                        random_state=42)
```

```python
np.round(x_train.describe(), 2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)  # fit menans learn the parameter and transform means apply it on data
```

```python
x_train_new = pd.DataFrame(x_train_sc , columns = x_train.columns)
```

```python
np.round(x_train_new.describe(), 2)
```

```python
mn = MinMaxScaler()
```
