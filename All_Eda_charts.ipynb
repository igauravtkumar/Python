1Ô∏è‚É£** Histogram üìä (For Single Numeric Column)**


```python
# ‚úÖ Used for: Understanding the distribution of a numerical column.
# ‚úÖ Best for: Checking skewness, normality, or outliers.
# üîπ X-axis: Values of the column
# üîπ Y-axis: Frequency (count of values in each bin)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'Age': [22, 25, 30, 30, 35, 40, 45, 50, 55, 60, 65, 70]})
plt.hist(df['Age'], bins=5, color='blue', edgecolor='black')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()
```

# **2  Box Plot üì¶ (For Outliers & Distribution)**

```python
# Used for: Detecting outliers, median, and quartiles in numerical data.
# Best for: Comparing distributions across different categories.
#  X-axis: Categories
#  Y-axis: Numeric values
```

```python
import seaborn as sns

df = pd.DataFrame({
    'Department': ['HR', 'HR', 'IT', 'IT', 'Sales', 'Sales'],
    'Salary': [30000, 35000, 50000, 55000, 40000, 42000]
})

sns.boxplot(x='Department', y='Salary', data=df)
plt.show()
```

# **Scatter Plot üìà (For Relationship Between Two Numeric Variables)**

```python
# ‚úÖ Used for: Checking the correlation between two numeric columns.
# ‚úÖ Best for: Identifying patterns, clusters, or trends.
# üîπ X-axis: One numeric variable
# üîπ Y-axis: Another numeric variable
```

```python
df = pd.DataFrame({'Ad_Spend': [100, 200, 300, 400, 500], 'Sales': [10, 20, 25, 40, 50]})

plt.scatter(df['Ad_Spend'], df['Sales'])
plt.xlabel("Advertising Spend")
plt.ylabel("Sales")
plt.title("Sales vs Advertising Spend")
plt.show()
```

# **Bar Chart üìä (For Categorical Data)**

```python
# ‚úÖ Used for: Comparing categories based on a numeric value.
# ‚úÖ Best for: Showing counts or percentages of different categories.
# üîπ X-axis: Categories
# üîπ Y-axis: Count or Numeric Value
```

```python
df = pd.DataFrame({'Category': ['Electronics', 'Clothing', 'Food'], 'Sales': [5000, 7000, 4000]})
plt.bar(df['Category'], df['Sales'], color=['blue', 'green', 'red'])
plt.xlabel("Category")
plt.ylabel("Sales")
plt.title("Sales by Category")
plt.show()
```

## **Pie Chart üçï (For Percentage Breakdown of Categories)**

```python
# ‚úÖ Used for: Showing proportion of different categories.
# ‚úÖ Best for: Visualizing market share, sales contribution, etc.
# üîπ Each slice represents a category's proportion in total.
```

```python
labels = ['Apple', 'Samsung', 'OnePlus']
sizes = [40, 35, 25]

plt.pie(sizes, labels=labels, autopct='%1.1f%%', colors=['red', 'blue', 'green'])
plt.title("Market Share Distribution")
plt.show()
```

# 6Ô∏è‚É£ Heatmap üî• (For Correlation Between Numeric Columns)

```python
# ‚úÖ Used for: Checking the relationship between multiple numeric columns.
# ‚úÖ Best for: Finding which variables are related (correlation).
```

```python
import seaborn as sns

df = pd.DataFrame({
    'Sales': [10, 20, 30, 40, 50],
    'Profit': [2, 5, 7, 10, 12],
    'Discount': [1, 2, 2, 3, 4]
})

sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

# Line Plot:

```python
# A line plot (or line chart) is used to visualize trends over time or show the relationship between two numeric variables.

# ‚úÖ When to Use a Line Plot?
# ‚úî Time Series Data (e.g., Sales over Months, Stock Prices over Time)
# ‚úî Trend Analysis (e.g., Temperature changes over days)
# ‚úî Comparing Multiple Series (e.g., Comparing Sales of different products)
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample Data: Sales Over Time
data = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'],
    'Sales': [1000, 1500, 1200, 1800, 2000, 2200]
})

# Plot Line Chart
plt.plot(data['Month'], data['Sales'], marker='o', linestyle='-', color='blue', label="Sales Trend")

# Labels and Title
plt.xlabel("Month")
plt.ylabel("Sales ($)")
plt.title("Monthly Sales Trend")
plt.legend()
plt.grid(True)

# Show Plot
plt.show()
```
