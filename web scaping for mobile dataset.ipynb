```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.shopsy.in/search?q=mobile%20under%2050000"
r = requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.text,'html')
print(soup)
```

```python
np = soup.find_all('div',class_="sc-c50e187b-0 bkNEtl")
print(np)
```

```python
name=[]
for i in np:
  names=i.text
  name.append(names)
print(name)
print(len(name))
```

```python
p = soup.find_all("div", class_ = "css-146c3p1 r-cqee49 r-1vgyyaa r-1rsjblm r-13hce6t")
price=[]
for i in p:
  prices = i.text
  price.append(prices)
print(price)
print(len(price))
```

```python
k = soup.find_all("div", class_="css-146c3p1 r-183gjk9 r-1vgyyaa r-1rsjblm r-11wrixw")
Offer=[]
for i in k:
  off = i.text
  Offer.append(off)
print(Offer)
print(len(Offer))
```

```python
l = soup.find_all('div', class_="css-146c3p1 r-dnmrzs r-1udh08x r-1udbk01 r-3s2u2q r-1iln25a r-1h7g6bg r-1et8rh5 r-1enofrn r-1f6r7vd r-l4nmg1 r-1kihuf0 r-9iso6")
review=[]
for i in l:
  reviews=i.text
  review.append(reviews)
print(review)
print(len(review))
```

```python
min_length = min(len(name), len(price), len(Offer),   len(review))

# Trim all lists to the minimum length
name = name[:min_length]
price = price[:min_length]
Offer = Offer[:min_length]
review = review[:min_length]
```

```python
df = pd.DataFrame({
    'Product Name': name,
    'Price': price,
    'Review': review,
    'Offer':Offer
})
print(df)
```

```python
df["Price"] = df["Price"].str.replace("₹", "")
df["Price"] = df["Price"].str.replace(",", "")
```

```python
df["Offer"] = df["Offer"].str.replace("% off", "")
df["Review"] = df["Review"].str.replace(",", "")
```

```python
df['Price']=df['Price'].astype(int)
```

```python
df['Review'] = df['Review'].astype(str).str.replace(r"[()]", "", regex=True)
```

```python
df['Review']=df['Review'].astype(int)
df['Offer']=df['Offer'].astype(int)
```

```python
df['Product_Name'] = df['Product Name'].str.split('(').str[0].str.strip()
df.drop('Product Name', axis=1, inplace=True)
cols = ['Product_Name'] + [col for col in df.columns if col != 'Product_Name']
df = df[cols]
```
