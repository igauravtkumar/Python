```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/covid_toy.csv')
df.head(3)
```

```python
df['gender']=df['gender'].map({'Male':1,'Female':0})
df['has_covid']=df['has_covid'].map({'Yes':1,'No':0})
df['cough']=df['cough'].map({'Mild':0,'Strong':1})
df['city']=df['city'].map({'Kolkata':0,'Bangalore':1,'Delhi':2,'Mumbai':3})
```

```python
df.sample(5)
```

```python
x = df.drop(columns=['has_covid'])
y=df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size = 0.2,
                                                       random_state=42)
```

```python
np.round(x_train.describe(),1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
x_train_sc = sc.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_sc)
```

```python
np.round(x_train_new.describe(),1)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mn = MinMaxScaler()
```

```python
x_train_mn = mn.fit_transform(x_train)
```

```python
x_train_new = pd.DataFrame(x_train_mn)
```

```python
np.round(x_train_new.describe(),1)
```
