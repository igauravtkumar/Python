```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df.head(5)
```

```python
sns.countplot(x=df["Region"])
```

```python
df['Region'].value_counts().plot(kind='pie',autopct = '%.2f')
```

```python
plt.figure(figsize=(8, 8))
iphone_sales_distribution = df.groupby("Region")["iPhone Sales (in million units)"].sum()
iphone_sales_distribution.plot(kind="pie", autopct="%.2f")
plt.title("iPhone Sales Distribution by Region")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
iphone_sales_distribution = df.groupby("Region")["iPad Sales (in million units)"].sum()
iphone_sales_distribution.plot(kind="pie", autopct="%.2f")
plt.title("iPad Sales Distribution by Region")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8, 8))
iphone_sales_distribution = df.groupby("Region")["Wearables (in million units)"].sum()
iphone_sales_distribution.plot(kind="pie", autopct="%.2f")
plt.title("Wearables Sales Distribution by Region")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="Region", y="Wearables (in million units)", data=df)
plt.title("Distribution of Wearables Sales by Region")
plt.ylabel("Wearables Sales (in million units)")
plt.xlabel("Region")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="Region", y="iPhone Sales (in million units)", data=df)
plt.title("Distribution of IPHONE Sales by Region")
plt.ylabel("iPhone Sales (in million units)")
plt.xlabel("Region")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="Region", y="iPad Sales (in million units)", data=df)
plt.title("Distribution of IPAD Sales by Region")
plt.ylabel("iPad Sales (in million units)")
plt.xlabel("Region")
plt.show()
```
