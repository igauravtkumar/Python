```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/insurance.csv')
df.head()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression , LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
x = df.drop(columns=['charges'])
y=df['charges']
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
categorical_feature = []
numerical_feature = []
def separate(x):
  for col in x.columns:
    if x[col].dtype == 'int64':
      numerical_feature.append(col)
    elif x[col].dtype == 'float64':
      numerical_feature.append(col)
    else:
      categorical_feature.append(col)
  return [categorical_feature,numerical_feature]

transform = FunctionTransformer(separate)

result = transform.transform(x)

print('Categorical : ',result[0])
print('Numerical : ',result[1])
```

```python
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
# create transform
numerical_transform = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='mean')),
    ('scale',StandardScaler())
])
categorical_transform = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('onehot',OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# combine transform
preprocessor = ColumnTransformer(
    transformers = [
        ('num',numerical_transform,numerical_feature),
        ('cat',categorical_transform,categorical_feature)
])
```

```python
clf = Pipeline(steps=[('preprocessor',preprocessor),
                      ('classifier',LinearRegression())])
clf.fit(x_train,y_train)
# Evaluate  the model
y_pred = clf.predict(x_test)

```

```python
from sklearn.metrics import r2_score
```

```python
acc = r2_score(y_test , y_pred)
acc
```
